## Linear Regression

### Hypothesis Function

$$
\hat{y} = b_0 + b_1 x
$$

### Loss Function (Mean Squared Error)

$$
J = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

### Objective

$$
\min_{b_0, b_1} J
$$

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

# Model
model = LinearRegression()
model.fit(X, y)

# Prediction
y_pred = model.predict(X)

print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
print("Predictions:", y_pred)
```

## Regression Evaluation Metrics

---

### Mean Squared Error (MSE)

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

---

### Root Mean Squared Error (RMSE)

$$
RMSE = \sqrt{ \frac{1}{n} \sum (y - \hat{y})^2 }
$$

---

### Mean Absolute Error (MAE)

$$
MAE = \frac{1}{n} \sum |y - \hat{y}|
$$

---

### R² Score (Coefficient of Determination)

$$
R^2 = 1 - \frac{\sum (y - \hat{y})^2}{\sum (y - \bar{y})^2}
$$

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

y_true = np.array([2, 4, 5, 4, 5])
y_pred = np.array([2.1, 3.9, 5.2, 3.8, 4.9])

mse = mean_squared_error(y_true, y_pred)

print("MSE:", mse)
print("RMSE:", np.sqrt(mse))
print("MAE:", mean_absolute_error(y_true, y_pred))
print("R2:", r2_score(y_true, y_pred))
```

## Linear Regression (Gradient Descent)

### Hypothesis

$$
\hat{y} = b_0 + b_1 x
$$

---

### Loss Function (Mean Squared Error)

$$
J = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

---

### Gradients

Derivative w.r.t intercept:

$$
\frac{\partial J}{\partial b_0} = -\frac{2}{n} \sum (y - \hat{y})
$$

Derivative w.r.t slope:

$$
\frac{\partial J}{\partial b_1} = -\frac{2}{n} \sum x (y - \hat{y})
$$

---

### Update Rules (Gradient Descent)

$$
b_0 = b_0 - \alpha \cdot \frac{\partial J}{\partial b_0}
$$

$$
b_1 = b_1 - \alpha \cdot \frac{\partial J}{\partial b_1}
$$

```python
class LinearRegression:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.lr = learning_rate
        self.epochs = epochs
        self.b0 = 0  # intercept
        self.b1 = 0  # slope

    def fit(self, X, y):
        n = len(X)

        for _ in range(self.epochs):
            y_pred = self.b0 + self.b1 * X

            # gradients
            db0 = (-2/n) * sum(y - y_pred)
            db1 = (-2/n) * sum(X * (y - y_pred))

            # update
            self.b0 -= self.lr * db0
            self.b1 -= self.lr * db1

    def predict(self, X):
        return self.b0 + self.b1 * X
```

```python
X = np.array([1, 2, 3, 4, 5])  # 1D
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()
model.fit(X, y)

pred = model.predict(X)

print(model.b0, model.b1)
print(pred)
```

```python
import numpy as np

def mse(y, y_pred):
    return np.mean((y - y_pred) ** 2)

def rmse(y, y_pred):
    return np.sqrt(np.mean((y - y_pred) ** 2))

def mae(y, y_pred):
    return np.mean(np.abs(y - y_pred))

def r2_score(y, y_pred):
    y_mean = np.mean(y)
    ss_total = np.sum((y - y_mean) ** 2)
    ss_res = np.sum((y - y_pred) ** 2)
    return 1 - (ss_res / ss_total)


# Example
y_true = np.array([2, 4, 5, 4, 5])
y_pred = np.array([2.1, 3.9, 5.2, 3.8, 4.9])

print("MSE:", mse(y_true, y_pred))
print("RMSE:", rmse(y_true, y_pred))
print("MAE:", mae(y_true, y_pred))
print("R2:", r2_score(y_true, y_pred))
```
